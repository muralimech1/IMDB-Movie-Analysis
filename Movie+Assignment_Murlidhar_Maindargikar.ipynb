-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here

#Code to create columns 'genre_1' and 'genre_2'. If only one genre is present then in both columns same value will be present
movies['genre_1'] = movies['genres'].str.split("|").apply(lambda x: x[0])
movies['genre_2'] = movies['genres'].str.split("|").apply(lambda x: x[0] if len(x)==1 else x[1])
movies
```

```python
#Grouping the dataframe based on genre_1 as primary column and genre_2 as secondary column
movies_by_segment = movies.groupby(['genre_1','genre_2'])
#Created movies_by_segment group object
```

```python
# Write your code for getting the 5 most popular combo of genres here

#Code to create PopGenre dataframe based on mean value of gross and sorting the same.
PopGenre = movies_by_segment['gross'].mean().sort_values(ascending = False).head(5)

print(PopGenre)

#Checkpoint 5: Confirmed that Family + Sci-Fi is most popular combo of genres.
```

**Checkpoint 5:** Well, as it turns out. `Family + Sci-Fi` is the most popular combo of genres out there!

Store this in a new data frame called `df_by_decade`.

```python
# Write your code for creating three new dataframes here

#Code to create Meryl_Streep dataframe where actor_1_name = Meryl Streep
Meryl_Streep = movies.loc[movies['actor_1_name'] == "Meryl Streep", :]
Meryl_Streep
```

```python
# Include all movies in which Leo_Caprio is the lead
#Code to create Leo_Caprio dataframe where actor_1_name = Leonardo DiCaprio
Leo_Caprio = movies.loc[movies['actor_1_name'] == "Leonardo DiCaprio", :]
Leo_Caprio
```

```python
# Include all movies in which Brad_Pitt is the lead
#Code to create Brad_Pitt dataframe where actor_1_name = Brad Pitt
Brad_Pitt = movies.loc[movies['actor_1_name'] == "Brad Pitt", :]
Brad_Pitt
```

```python
# Write your code for combining the three dataframes here

#Creating Combined dataframe by using Meryl_Streep, Leo_Caprio, Brad_Pitt dataframes
Combined = pd.concat([Meryl_Streep, Leo_Caprio, Brad_Pitt])
Combined
```

```python
# Write your code for grouping the combined dataframe here  

#Creating group by object segment by actor nased on 'actor_1_name'
segment_by_actor = Combined.groupby('actor_1_name')
```

```python
# Write the code for finding the mean of critic reviews and audience reviews here

#Code to find mean of critic reviews and audience reviews for each actor
print(segment_by_actor['num_critic_for_reviews','num_user_for_reviews'].mean().
      sort_values(by = ['num_critic_for_reviews','num_user_for_reviews'],ascending = False))

#Checkpoint 6:Leonardo is the at the top position in both list
```

**Checkpoint 6:** `Leonardo` has aced both the lists!

```python
# Write the code for calculating decade here

#Code to create 'decade' column
movies['decade'] = movies['title_year'].apply(lambda x: int(x- x%10))

#Sorting dataframe based on values in decase column
movies = movies.sort_values(by = 'decade')

movies
```

```python
# Write your code for creating the data frame df_by_decade here 

#Code to create dataframe containing decade and sum of number of voted users
df_by_decade = movies.groupby('decade').sum()['num_voted_users']
pd.DataFrame(df_by_decade)
```

```python
# Write your code for plotting number of voted users vs decade

#Code to plot the barplot based on decade and number of voted users
plt.figure()
df_by_decade.plot.bar()
plt.ylabel('number of voted users')
#Code to change the scale of Y axis value to have a better visualization of the plot
plt.yscale('log')
plt.title('Plot for number of voted users vs decade')
plt.show()

#As per observation, we can see that number of users are increasing over the decades.
```
